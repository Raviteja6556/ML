```python
df=pd.read_excel("Book3.xlsx")
df.head(15)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Create scatter plot
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='Label', y='Value', hue='Sep', palette='deep')

# Add line plot
for Sep in df['Sep'].unique():
    subset = df[df['Sep'] == Sep]
    plt.plot(subset['Label'], subset['Value'], label=Sep)

plt.xticks(ticks=range(0, 35, 2))
plt.yticks(ticks=range(0, 201, 10))
plt.xlabel('Label')
plt.ylabel('Value')
plt.title('Scatter Plot with Line Graph')
plt.legend(title='Sep')
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Assuming df is your existing DataFrame
# Create scatter plot with line plot
fig = px.scatter(df, x='Label', y='Value', color='Sep', title='Scatter Plot with Line Graph')

# Add line plot
for sep in df['Sep'].unique():
    subset = df[df['Sep'] == sep]
    fig.add_trace(go.Scatter(x=subset['Label'], y=subset['Value'], mode='lines', name=sep))

# Customize the layout
fig.update_layout(
    xaxis=dict(tickmode='linear', tick0=0, dtick=2),
    yaxis=dict(tickmode='linear', tick0=0, dtick=10),
    xaxis_title='Label',
    yaxis_title='Value',
    legend_title='Sep'
)
fig.show()
```
